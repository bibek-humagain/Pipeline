```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.datasets import load_boston
```

```python
DataX,DataY = load_boston(return_X_y=True)
DataY = np.expand_dims(DataY,axis=1)
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.neural_network import MLPRegressor
```

```python
from sklearn.metrics import r2_score
```

```python
from sklearn.preprocessing import MinMaxScaler
scalerX = MinMaxScaler()
scalerY = MinMaxScaler()
scalerX.fit(DataX)
scalerY.fit(DataY)

from sklearn.model_selection import train_test_split
X,TestX,Y,TestY = train_test_split(DataX,DataY,test_size=0.1)
```

```python
batch_size = 30
passes = 100
Models = []
Performances = []
for beta in [0.0001,0.001, 0.01, 0.1,1]:
    for lamb in [0.0001,0.001,0.01,0.1,0.3,0.7,0.9]:
        model = SGDRegressor(loss = 'squared_loss',
                             penalty='elasticnet',
                             alpha = beta,l1_ratio=lamb,
                             learning_rate='constant',
                             eta0=1e-2,
                             early_stopping=False,
                             verbose=False)


        Losses = []
        for _ in range(passes):
            for i in range(0,X.shape[0],batch_size):

                model.partial_fit(scalerX.transform(X[i:i+batch_size]),
                                  scalerY.transform(Y[i:i+batch_size]).flatten())
                Losses.append(model.score(scalerX.transform(X[i:i+batch_size]),
                              scalerY.transform(Y[i:i+batch_size]).flatten()))

        plt.plot(Losses)
        
        Models.append(model)
        
        Performances.append(r2_score(scalerY.transform(TestY),
                                     y_pred=model.predict(scalerX.transform(TestX))))
        print('Model \n \n {} Performance :{}'.format(model,Performances[-1]))
        
```

```python
model = Models[np.argmax(Performances)]
```

```python
plt.scatter(scalerY.transform(TestY),model.predict(scalerX.transform(TestX)))
plt.plot(scalerY.transform(TestY),scalerY.transform(TestY),'-k')
```
