```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## simulate some data
X1 = np.random.randn(1000,2)+np.array([[3,3]])
```

```python
X2 = np.random.randn(500,2)+np.array([[-1,-1]])
```

```python
## plot the data
plt.scatter(X2[:,0],X2[:,1])
```

```python
X = np.vstack((X1,X2))
Y = np.vstack((np.ones((X1.shape[0],1)),np.zeros((X2.shape[0],1))))
```

```python
X = np.hstack((np.ones((X.shape[0],1)),X))
```

```python
def sigmoid(I):
    return 1/(1+np.exp(-I))
```

```python
def BCEC(p,y):
    return -np.sum(y*np.log(p)+(1-y)*np.log(1-p))
```

```python
X.shape
```

```python
## gradient descent
# initialize weights
w = np.random.randn(3,1)
epochs = 1000
eta = 1e-3
Losses = []

for i in range(epochs):
    Loss = BCEC(sigmoid(X@w),Y)
    Losses.append(Loss)
    print("iteration : {} Loss : {}".format(i+1,Loss))
    w -= eta*X.T@(sigmoid(X@w)-Y)
```

```python
v1 = np.random.randn(1000,1)*2
v2 = (-w[0]-v1*w[1])/w[2]
```
