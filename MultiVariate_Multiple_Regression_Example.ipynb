### Let's import libraries and create some synthetic data

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# create an (N,2) matrix of random numbers, add one column of ones to the left
N = 200
X = np.hstack((np.ones((N,1)),np.random.randn(N,1)*300,np.random.randn(N,1)*100+50))
X.shape
```

```python
# create a vector of true weights 
w_true = np.array([[3.,7.,9.],[1.,2.,3.],[4.,5.,6.]]).reshape(3,3)
# generate true y with some noise
y = X.dot(w_true) + np.random.randn(N,3)
print("w_true.shape = {}".format(w_true.shape))
print("y.shape = {}".format(y.shape))
```

### 3D scatter plot

```python
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(X[:,1],X[:,2],y[:,0])
```

### Find predicted weight matrix

```python
w = np.linalg.solve(X.T.dot(X),X.T.dot(y))
# Or w = np.linalg.inve(X.T.dot(X)).dot(X.T.dot(y))
print("w.shape = {}".format(w.shape))
```

### Get predicted y

```python
yhat = X.dot(w)
print("yhat.shape = {}".format(yhat.shape))
print(w)
```

### Get R-squared. It's a different formula now!

```python
Rsq = 1- (np.trace((y-yhat).T.dot(y-yhat)))/(np.trace((y-y.mean()).T.dot(y-y.mean())))
print("Rsquared = {}".format(Rsq))
```
