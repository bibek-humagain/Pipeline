```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 300
X = np.hstack((np.ones((N,1)),np.random.rand(N,1)))
b = np.array([[2],[3]])
Y = X@b+np.random.randn(N,1)*0.1
```

```python
Y[210:215] = Y[210:215]+np.random.rand(5,1)*10
```

```python
plt.scatter(X[:,[1]],Y)
```

```python
w = np.linalg.inv(X.T@X)@X.T@Y
```

```python
plt.scatter(X[:,[1]],Y,s=2)
plt.plot(X[:,[1]],X@w,'-r')
```

```python
Rsq = 1- ((Y-X@w).T@(Y-X@w))/((Y-Y.mean()).T@(Y-Y.mean()))
Rsq
```

```python
# closed form solution of L1 regularization
lamb = 0.001
eta = 1e-3
epochs=4000
Errors=np.empty(epochs)
for i in range(epochs):
    w-= eta*X.T@(X@w-Y)+lamb*w
Rsqs = 1- ((Y-X@w).T@(Y-X@w))/((Y-Y.mean()).T@(Y-Y.mean()))
Rsqs
```
