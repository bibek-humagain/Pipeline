```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Xs = np.random.randn(1000,2)*20
Ys= np.random.randn(1000,2)
Donut = np.array(list(filter(lambda x: x[0]**2 + x[1]**2 >= 6**3.5, Xs)))
```

```python
plt.scatter(Donut[:,0], Donut[:,1])
plt.scatter(Ys[:,0], Ys[:,1])
Y= np.vstack((np.ones((Ys.shape[0],1)), np.zeros((Donut.shape[0],1))))
X= np.vstack((Ys, Donut))
Phi = np.hstack((np.ones((X.shape[0],1)),X,np.sqrt(X[:,[0]]**2+X[:,[1]]**2)))
```

```python
Y.shape
```

```python
def sigmoid(I):
    return 1/(1+np.exp(-I))
```

```python
def BCEC(p,y):
    return -np.sum(y*np.log(p)+(1-y)*np.log(1-p))
```

```python
## gradient descent
# initialize weights
w = np.random.randn(4,1)
epochs = 1000
eta = 1e-3
Losses = []

for i in range(epochs):
    Loss = BCEC(sigmoid(Phi@w),Y)
    Losses.append(Loss)
    print("iteration : {} Loss : {}".format(i+1,Loss))
    w -= eta*Phi.T@(sigmoid(Phi@w)-Y)
```

```python
simX = (np.random.rand(20000,2)-0.5)*(X.max(axis=0)-X.min(axis=0))
```

```python
simPhi = np.hstack((np.ones((simX.shape[0],1)),simX,np.sqrt(simX[:,[0]]**2+simX[:,[1]]**2)))
```

```python
DB =np.array(list(filter(lambda x: x@w>=-0.4 and x@w<=0.3, simPhi)))
```
