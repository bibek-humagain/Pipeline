```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 500
K=5
b = 10
X = np.hstack((np.ones((K*N,1)),
               np.vstack((np.random.randn(N,2)+np.random.randn(1,2)*b,
                          np.random.randn(N,2)+np.random.randn(1,2)*b,
                          np.random.randn(N,2)+np.random.randn(1,2)*b,
                          np.random.randn(N,2)+np.random.randn(1,2)*b,
                          np.random.randn(N,2)+np.random.randn(1,2)*b))))
```

```python
plt.scatter(X[:,1],X[:,2])
```

```python
Y = np.vstack(([[1,0,0,0,0]]*N,
               [[0,1,0,0,0]]*N,
               [[0,0,1,0,0]]*N,
               [[0,0,0,1,0]]*N,
               [[0,0,0,0,1]]*N))
```

```python
w = np.random.randn(3,5)
```

```python
X[:,1:] = (X[:,1:]-X[:,1:].min(axis=0,keepdims=1))/(X[:,1:].max(axis=0,keepdims=1)-X[:,1:].min(axis=0,keepdims=1))
```

```python
X.shape
```

```python
def softmax(h):
    num = np.exp(h)
    return num/np.sum(num,axis=1,keepdims=1)

def GCEC(P,Y):
    return np.sum(Y*np.log(P))
```

```python
eta = 1e-2
epochs = 2000

for i in range(epochs):
    print('Iteration:{}Loss:{}'.
          format(i+1,GCEC(softmax(X@w),Y)))
    w-=eta*X.T@(softmax(X@w)-Y)
```

```python
# find maximum value in every row of Phat 
# to find which class every row belongs to
np.argmax(softmax(X@w),axis=1)
```

```python
plt.scatter(X[:,1],
            X[:,2],
            c=np.argmax(softmax(X@w),axis=1))
```

```python
simX = np.hstack((np.ones((1000,1)),np.random.rand(1000,2)))
```

```python
plt.scatter(simX[:,1],simX[:,2],c=np.argmax(softmax(simX@w),axis=1))
```
