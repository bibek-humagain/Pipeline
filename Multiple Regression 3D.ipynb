```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# declare number of observations
N = 500
# simulate data(predictors) and add a dummy column of ones to incroporate bias
X = np.hstack((
    np.ones((N,1)),
    np.random.rand(N,2)*1000))
```

```python
# simulate weight matrix 3 by 1 (one of them is the bias)
b = np.array([[50.],[2.2],[3.7]])
```

```python
# simulate the response variables with a matrix product and addition of noise
Y = X@b+np.random.randn(500,1)*500
```

```python
# print dimensions to see if everything lines up
print('dimensions of X={}, b={}, Y={}'.format(X.shape,b.shape,Y.shape))
```

```python
# 3D plot to see what our simulated plane looks like
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,1],X[:,2],Y,s=4,c='k')
```

```python
# one liner for regression using the vectorized OLS solution/ Moore Penrose PseudoInverse/ Normal Equation
w = np.linalg.inv(X.T@X)@X.T@Y
```

```python
print('w = {} \n\n whereas \n\n b = {}'.format(w.flatten(),b.flatten()))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,1],X[:,2],Y,s=4,c='k')
ax.plot(X[:,1],X[:,2],(X@w).flatten(),'-r')
```
