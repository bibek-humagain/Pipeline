```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simulate some data
X  = np.random.randn(100,2)@np.array([[1.2, 0.8],[0.8, 1.2]])+[[20, 100]]

# Normalize your data as eigendecomposition 
# assumes that there was no translation in the transformation
X = (X-X.mean(axis=0))/X.std(axis=0)
```

```python
plt.scatter(X[:,0],X[:,1],s=3)
plt.show()
```

```python
C = 1
lamb,V = np.linalg.eig(X.T@X)
Comps = V[:,np.argsort(lamb)[::-1][:C]]
CompOrth = V[:,np.argsort(lamb)[::-1][C]]
Xred = X@Comps
```
