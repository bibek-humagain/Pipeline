```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X= np.linspace(-100,100,100).reshape(100,1)
```

```python
Y= X@np.array([[3]])+np.array([100])+np.random.randn(100,1)*25

```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,Y)
plt.title('Regression Toy Data Set')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
W0 = (Y.mean()*np.mean(X**2)-np.mean(X)*np.mean(X*Y))/(np.mean(X**2)-X.mean()**2)
```

```python
W1= (np.mean(X*Y)-np.mean(X)*Y.mean())/(np.mean(X**2)-X.mean()**2)
```

```python
yhat = W0 + X@np.array([[W1]])
```

```python
plt.plot(X, yhat)
plt.scatter(X,Y)
```

```python
R = 1 -((Y - yhat).T@(Y-yhat))/((Y-Y.mean()).T@(Y-Y.mean())) 
```

```python
print(R)
```
