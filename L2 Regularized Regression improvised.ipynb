```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 300
X = np.hstack((np.ones((N,1)),np.random.rand(N,1)))
b = np.array([[2],[3]])
Y = X@b+np.random.randn(N,1)*0.1
```

```python
plt.scatter(X[:,[1]],Y)
```

```python
Y[210:215] = Y[210:215]+np.random.rand(5,1)*10
```

```python
plt.scatter(X[:,[1]],Y)
```

```python
w = np.linalg.inv(X.T@X)@X.T@Y
```

```python
plt.scatter(X[:,[1]],Y,s=2)
plt.plot(X[:,[1]],X@w,'-r')
```

```python
# closed form solution of L2 regularization 
# with different reg parameters for each weight
lamb = np.array([[5],[2]])
wreg = np.linalg.inv(X.T@X+np.eye(b.shape[0])@lamb)@X.T@Y

# closed form OG regularization
# lamb = 5
# wreg = np.linalg.inv(X.T@X+lamb*np.eye(b.shape[0]))@X.T@Y
```

```python
plt.scatter(X[:,[1]],Y,s=2)
plt.plot(X[:,[1]],X@wreg,'-r')
```

```python
Rsq = 1- ((Y-X@w).T@(Y-X@w))/((Y-Y.mean()).T@(Y-Y.mean()))
RsqReg = 1- ((Y-X@wreg).T@(Y-X@wreg))/((Y-Y.mean()).T@(Y-Y.mean()))
print('Rsq={}, regularized Rsq={}'.format(Rsq[0][0],RsqReg[0][0]))
```
